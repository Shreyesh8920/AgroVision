```python
import pickle
import pandas as pd

with open('gb_model_maize.pkl', 'rb') as f:
    gb_model = pickle.load(f)

with open('label_encoders_maize.pkl', 'rb') as f:
    labelencoders = pickle.load(f)
```

```python
for col, le in labelencoders.items():
    print(f"Mapping for {col}:")
    for i, cls in enumerate(le.classes_):
        print(f"  {cls} -> {i}")
```

```python

def predict_yield(input_data):
    input_df = pd.DataFrame(input_data)
    for col, le in labelencoders.items():
        if col in input_df.columns:
            input_df[col] = le.transform(input_df[col])
    
    prediction = gb_model.predict(input_df)
    return prediction[0]
```

```python
input_data = {
    'Region': ['North'],
    'Soil_Type': ['Silt'],
    'Rainfall_mm': [0],#689.113795
    'Temperature_Celsius': [25.801389],
    'Fertilizer_Used': [True],#true
    'Irrigation_Used': [True],#true
    'N': [246.886589],
    'P': [15.363449],
    'K': [173.453087],
    'ph': [7.410310]
}

predicted_yield = predict_yield(input_data)
print("Predicted Yield:", predicted_yield)




```
