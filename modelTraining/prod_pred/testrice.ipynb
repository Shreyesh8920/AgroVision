```python
import pickle
import pandas as pd
with open('gb_model_rice.pkl', 'rb') as f:
    gb_model = pickle.load(f)

with open('label_encoders_rice.pkl', 'rb') as f:
    labelencoders = pickle.load(f)
```

```python
for col, le in labelencoders.items():
    print(f"Mapping for {col}:")
    for i, cls in enumerate(le.classes_):
        print(f"  {cls} -> {i}")
```

```python


def predict_yield(input_data):
    input_df = pd.DataFrame(input_data)

    for col, le in labelencoders.items():
        if col in input_df.columns:
            input_df[col] = le.transform(input_df[col])
    
    prediction = gb_model.predict(input_df)
    return prediction[0]
```

```python
region_le = labelencoders['Region']

for i, cls in enumerate(region_le.classes_):
    print(f"{cls} -> {i}")

key = 'South'
encoded_val = region_le.transform([key])[0]
print(f"Encoded value of {key}: {encoded_val}")
```

```python
sample_input = {
    'Region': ['West'],
    'Soil_Type': ['Chalky'],
    'Rainfall_mm': [590.309080],
    'Temperature_Celsius': [30.241923],
    'Fertilizer_Used': [True],
    'Irrigation_Used': [True],
    'N': [145.558201],
    'P': [6.911538],
    'K': [158.498665],
    'ph': [8.287763]
}


predicted_yield = predict_yield(sample_input)
print("Predicted Yield:", predicted_yield)


```
