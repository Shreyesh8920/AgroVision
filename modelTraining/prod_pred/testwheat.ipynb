```python
import pickle
import pandas as pd
with open('gb_model_wheat.pkl', 'rb') as f:
    gb_model = pickle.load(f)
with open('label_encoders_wheat.pkl', 'rb') as f:
    labelencoders = pickle.load(f)
```

```python
for col, le in labelencoders.items():
    print(f"Mapping for {col}:")
    for i, cls in enumerate(le.classes_):
        print(f"  {cls} -> {i}")
```

```python

def predict_yield(input_data):
    input_df = pd.DataFrame(input_data)

    for col, le in labelencoders.items():
        if col in input_df.columns:
            input_df[col] = le.transform(input_df[col])
    
    prediction = gb_model.predict(input_df)
    return prediction[0]
```

```python
input_data = {
    'Region': ['North'],           # string
    'Soil_Type': ['Silt'],         # string
    'Rainfall_mm': [820.460446], # annual rainfall -> unit -> mm
    'Temperature_Celsius': [31.003539],
    'Fertilizer_Used': [False],   # Boolean type, not string 'False'
    'Irrigation_Used': [False],   # Boolean type, not string 'False'
    'N': [157.510061], # unit-> kg/ha
    'P': [16.875108], # unit-> kg/ha
    'K': [152.58378], # unit-> kg/ha
    'ph': [7.710211], # ph scale
}

predicted_yield = predict_yield(input_data)
print("Predicted Yield:", predicted_yield)



# Region -> dropwown -> North, East, West, South
# Soil_type -> dropdown ->  Chalky, Clay, Loam, Peaty, Sandy, Silt 
# Fertilizer_Used, Irregation_Used -> True, False
```
